```
df_titanic=pd.read_csv('TitanicClean.csv')
df_titanic.drop('Unnamed: 0',axis=1,inplace=True)
df_titanic
```

```
d_fam=df_titanic[['Parch','SibSp','Survived']]
d_fam['Fsize']=d_fam['Parch']+d_fam['SibSp']+1
d_fam.drop('Parch',inplace=True,axis=1)
d_fam.drop('SibSp',inplace=True,axis=1)
d_fam.drop('Survived',inplace=True,axis=1)
```

```
df_titanic['Parch']=d_fam['Fsize']
df_titanic.rename(columns={'Parch':'Fsize'},inplace=True)
df_titanic.drop('SibSp',inplace=True,axis=1)
```

```
df_titanic
```

```
df_titanic['Died'].replace(0,'Alive',inplace=True)
df_titanic['Died'].replace(1,'Dead',inplace=True)
```
